## 1. La Ecuación Fundamental del Orificio

Cualquier correlación parte de la ecuación de Bernoulli modificada.

La caída de presión ($\Delta P$) se relaciona con el caudal mediante un coeficiente de descarga

($C_d$):$$Q = C_d \cdot E \cdot A_o \cdot \sqrt{\frac{2\Delta P}{\rho}}$$

Donde:
* $C_d$: Coeficiente de descarga (ajusta el flujo real vs. el ideal).

* $E$: Factor de velocidad de aproximación: $1/\sqrt{1-\beta^4}$.

* $\beta$: Relación de diámetros ($d_{orificio} / D_{tubería}$).

## Correlaciones para el Coeficiente de Descarga
($C_d$)

El factor de fricción en un orificio no se calcula como el de una tubería (Fanning o Moody), sino que se "absorbe" dentro del $C_d$.

Estas son las ecuaciones más utilizadas:

**A. Reader-Harris/Gallagher (ISO 5167)**

Es el estándar internacional actual (adoptado por la norma ISO 5167-2:2003).
Es una ecuación empírica extremadamente compleja que considera el número de Reynolds ($Re$) y la ubicación de las tomas de presión.

Es la ecuación más robusta. Para una placa de orificio, el coeficiente de descarga $C_d$ se calcula como:
$$C_d = 0.5961 + 0.0261\beta^2 - 0.216\beta^8 + 0.000521\left(\frac{10^6\beta}{Re_D}\right)^{0.7} + (0.0188 + 0.0063A)\beta^{3.5}\left(\frac{10^6}{Re_D}\right)^{0.3} + (0.043 + 0.080e^{-10L_1} - 0.123e^{-7L_1})(1 - 0.11A)\frac{\beta^4}{1-\beta^4} - 0.031(M_2' - 0.8M_2'^{1.1})\beta^{1.3}$$

Donde los parámetros de ubicación de toma ($L_1, L_2$) son:Tomas en la brida (Flange Taps):

$L_1 = L_2 = 25.4 / D$ (con $D$ en mm).Tomas en la esquina (Corner Taps):

$L_1 = L_2 = 0$.Tomas $D$ y

$D/2$: $L_1 = 1$ y $L_2 = 0.47$.

$A$ (Factor de rugosidad): $A = \left(\frac{19000\beta}{Re_D}\right)^{0.8}$

Uso: Es la más precisa para transferencia de custodia y simulaciones de alta fidelidad.

Variables: Depende de $\beta$, $Re_D$, y términos adicionales para tomas en la brida (flange taps), radio o esquina.

**B. Stolz**

Utilizada ampliamente antes de la estandarización total de la ISO 5167, es una forma simplificada de la anterior. Todavía aparece en códigos de legado y simuladores de procesos antiguos.

**C. Miller (AGA 3)**

Muy común en la industria del gas natural en América. La ecuación de la AGA Report No. 3 es similar a la ISO, pero con ajustes específicos para la compresibilidad de gases.
